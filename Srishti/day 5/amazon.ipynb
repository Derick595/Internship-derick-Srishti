```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path=r"C:\Users\DERICK\OneDrive\Documents\Intership\Srishti\day 5\amazon.csv"
df = pd.read_csv(file_path)
display(df.head())
display(df.info())
```

```python
df.shape
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(df['actual_price'].dropna(), fill=True, color='skyblue')
plt.title('Actual Price Distribution (KDE)')
plt.xlabel('Actual Price (₹)')
plt.ylabel('Density')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='discounted_price', kde=True, bins=30, color='orange')
plt.title('Discounted Price Distribution')
plt.xlabel('Discounted Price (₹)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

price_bins = [0, 500, 1000, 2000, 5000, df['actual_price'].max()]
df['price_range'] = pd.cut(df['actual_price'], bins=price_bins, labels=['0-500', '500-1000', '1000-2000', '2000-5000', '5000+'])
price_range_counts = df['price_range'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.barplot(x=price_range_counts.index, y=price_range_counts.values, palette='magma')
plt.title('Actual Price Range Distribution')
plt.xlabel('Price Range (₹)')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.show()
```

## Price Distribution Overview

```python
category_counts = df['category'].value_counts().head(10)
display(category_counts)

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=category_counts.values, y=category_counts.index, palette='Blues_r')
plt.title('Top 10 Product Categories by Count')
plt.xlabel('Number of Products')
plt.ylabel('Category')
for i, v in enumerate(category_counts.values):
    ax.text(v + 5, i, str(v), color='black', va='center', fontweight='bold')
plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 7))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Top 10 Categories Share')
plt.axis('equal')
plt.show()
```

## Category Distribution

```python
missing = df.isnull().sum()
display(missing)

plt.figure(figsize=(10, 5))
ax = missing.plot(kind='bar', color='forestgreen')
plt.title('Missing Values per Column')
plt.xlabel('Columns')
plt.ylabel('Number of Missing Values')
plt.xticks(rotation=45)
for index, value in enumerate(missing):
    ax.text(index, value + 0.1, str(value), ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Missing Value Analysis

```python
# Ensure numeric columns are properly converted
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].astype(str).str.rstrip('%'), errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', ''), errors='coerce')
df['discounted_price'] = pd.to_numeric(df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', ''), errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'].astype(str).str.replace(',', ''), errors='coerce')

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

# Create correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={'label': 'Correlation'})
plt.title('Correlation Matrix of Numeric Features')
plt.tight_layout()
plt.show()
```

## Correlation Heatmap

```python
# Visualize discount percentage distribution
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='discount_percentage', kde=True, bins=30, color='orange')
plt.title('Distribution of Discount Percentages')
plt.xlabel('Discount Percentage (%)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Discount Percentage Distribution

```python
# Average rating by category
# Convert rating to numeric first
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
top_categories = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
ax = top_categories.plot(kind='barh', color='teal')
plt.title('Top 10 Categories by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Category')
# Zoom in on the rating range to show differences better
ax.set_xlim(top_categories.min() - 0.1, top_categories.max() + 0.1)
plt.tight_layout()
plt.show()
```

## Top Categories by Average Rating

```python
# Scatter plot: Actual Price vs Rating
plt.figure(figsize=(12, 6))
plt.scatter(df['actual_price'], df['rating'], alpha=0.5, color='steelblue')
plt.title('Relationship between Actual Price and Rating')
plt.xlabel('Actual Price (₹)')
plt.ylabel('Rating')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Relationship: Price vs Rating

```python
# Compare discounted vs actual prices
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Discounted price distribution
axes[0].hist(df['discounted_price'], bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of Discounted Prices')
axes[0].set_xlabel('Discounted Price (₹)')
axes[0].set_ylabel('Frequency')

# Actual price distribution
axes[1].hist(df['actual_price'], bins=30, color='lightcoral', edgecolor='black')
axes[1].set_title('Distribution of Actual Prices')
axes[1].set_xlabel('Actual Price (₹)')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
